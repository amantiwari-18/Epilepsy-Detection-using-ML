```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score


print("hiiii")

ep_data = pd.read_csv("/content/drive/MyDrive/data.csv")

ep_data.head()



ep_data.info()

ep_data.isnull().sum()

ep_data.describe()

ep_data['output'].value_counts()

has =ep_data[ep_data.output == "NO"]
no = ep_data[ep_data.output == "YES"]

print(has.shape)
print(no.shape)

has=has.sample(n=200)

no=no.sample(n=200)

print(has.shape)
print(no.shape)

new_dataset = pd.concat([has,no], axis = 0)

new_dataset.groupby('output').mean()



X = new_dataset.drop(columns=['output'], axis=1)
Y = new_dataset['output']

print(X)

print(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

print(X.shape, X_train.shape, X_test.shape)

scaler = StandardScaler()

scaler.fit(X_train)

print(X_train)
print("**************TRANING*************")


model = svm.SVC(kernel='linear')

model.fit(X_train, Y_train)

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)

print('Accuracy score of training data : ', training_data_accuracy)
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)

print('Accuracy score of test data : ', test_data_accuracy)
```
